```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cab_data = pd.read_csv("Cab.csv")
customer_data = pd.read_csv("Customer.csv")
transaction_data = pd.read_csv("Transaction.csv")
city_data = pd.read_csv("City.csv")
```

```python
cab_data = pd.read_csv("Cab.csv")
cab_data["Date of Travel"] = pd.to_datetime(cab_data["Date of Travel"], unit="D", origin="1899-12-30")
print(cab_data.head())
cab_data.head()
```

```python
city = pd.read_csv("City.csv")
city.head()
```

```python
customer_transaction =pd.merge(customer_data, transaction_data, on="Customer ID")
customer_transaction
```

```python
cab_data = pd.merge(cab_data, city_data, on="City")
cab_data
```

```python
cab_data = pd.merge(cab_data, city_data, on="City")
```

```python
main_data = pd.merge(cab_data, customer_transaction, on="Transaction ID")
main_data
```

```python
main_data.drop_duplicates(inplace=True)
```

```python
print("\nAna NA values ​​in the data set:")
print(main_data.isnull().sum())
```

```python
main_data
```

```python
sorted_data = main_data.sort_values(by="Date of Travel")
```

```python
sorted_data
```

```python
city_total_rides = sorted_data.groupby('City')['KM Travelled'].count().reset_index()
city_total_rides.rename(columns={'KM Travelled': 'Total Rides'}, inplace=True)
        
```

```python
city_total_rides = pd.merge(city_total_rides, city_data[['City', 'Population']], on='City', how='left')
```

```python
city_total_rides['Population'] = city_total_rides['Population'].str.replace(',', '').astype(int)
```

```python
city_total_rides['Rides per 1000 People'] = (city_total_rides['Total Rides'] / city_total_rides['Population'])
```

```python
top_cities = city_total_rides.nlargest(5, 'Rides per 1000 People')
```

```python
print(top_cities)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(top_cities['City'], top_cities['Rides per 1000 People'], color='skyblue')
plt.title('Cities with the Highest Taxi Usage Density')
plt.xlabel('City')
plt.ylabel('Taxi Usage (Per 1000 People)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
cab_data = pd.read_csv("Cab.csv")
```

```python
avg_price_charged = cab_data.groupby('Price Charged')['Price Charged'].mean()
avg_cost_of_trip = cab_data.groupby('Cost of Trip')['Cost of Trip'].mean()
```

```python
plt.figure(figsize=(10, 6))

plt.plot(avg_price_charged, color='blue', label='Average Wage Collected')
plt.plot(avg_cost_of_trip, color='red', label='Average Trip Cost')
plt.title('Average Fare and Trip Cost per Customer')
plt.xlabel('User ID')
plt.ylabel('Average Fee/Cost')
plt.legend()
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(sorted_data['Cost of Trip'], sorted_data['Price Charged'], color='blue', alpha=0.5)
plt.title('Trip Cost and Fee Collected Distribution')
plt.xlabel('Trip Cost')
plt.ylabel('Fee Collected')
plt.grid(True)
plt.show()
```

```python
total_revenue = sorted_data['Price Charged'].sum()
total_cost = sorted_data['Cost of Trip'].sum()

labels = ['Total Revenue', 'Total Cost']
sizes = [total_revenue, total_cost]
colors = ['lightblue', 'lightcoral']
explode = (0.1, 0)

plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Total Revenue and Total Cost Distribution')
plt.axis('equal')
plt.show()
```

```python
# Converting "Date of Travel" column to date format
sorted_data['Date of Travel'] = pd.to_datetime(sorted_data['Date of Travel'])
```

```python
# Create a new column by taking the year and month information from the "Date of Travel" column
sorted_data['Year-Month'] = sorted_data['Date of Travel'].dt.to_period('M')
```

```python
# Calculate the number of trips each company takes in a given time period
company_travel_count = sorted_data.groupby(['Company', 'Year-Month']).size().unstack(fill_value=0)
```

```python
#visualization
plt.figure(figsize=(12,6))
company_travel_count.plot(kind='bar', stacked=True, colormap='viridis')
plt.title('Number of Trips of Companies on a Year-Month Basis')
plt.xlabel('Year-month')
plt.ylabel('Number of Trips')
plt.xticks(rotation=45)
plt.legend(title='Company', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```

```python
pink_cab_data = sorted_data[cab_data['Company'] == 'Pink Cab']
yellow_cab_data = sorted_data[cab_data['Company'] == 'Yellow Cab']
```

```python
#Calculate trip numbers for "Pink Cab" and "Yellow Cab" on a year-month basis
pink_cab_travel_count = pink_cab_data.groupby('Year-Month').size()
yellow_cab_travel_count = yellow_cab_data.groupby('Year-Month').size()
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
pink_cab_travel_count.plot(kind='bar', color='deeppink')
plt.title('Pink Cab  Number of Trips on Year-Month Basis')
plt.xlabel('Year - month')
plt.ylabel('Number of Trips')
plt.xticks(rotation=45)

plt.subplot(2, 1, 2)
yellow_cab_travel_count.plot(kind='bar', color='gold')
plt.title('Yellow Cab Number of Trips on Year-Month Basis')
plt.xlabel('Year - month')
plt.ylabel('Number of Trips')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Number of trips by gender and age
gender_travel_count = main_data.groupby('Gender').size()
age_travel_count = main_data.groupby('Age').size()
```

```python
# Average fees by payment method
payment_mode_avg_price = main_data.groupby('Payment_Mode')['Price Charged'].mean()
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
gender_travel_count.plot(kind='bar', color='skyblue')
plt.title('Number of Trips by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Trips')

plt.subplot(2, 1, 2)
age_travel_count.plot(kind='bar', color='lightgreen')
plt.title('Number of Trips by Age')
plt.xlabel('Age')
plt.ylabel('Number of Trips')

plt.tight_layout()
plt.show()
```

```python
# Average fees by payment method
plt.figure(figsize=(8, 6))
payment_mode_avg_price.plot(kind='bar', color='red')
plt.title('Average Fees by Payment Method')
plt.xlabel('Payment method')
plt.ylabel('Average Wage')
plt.xticks(rotation=0)
plt.show()
```
